```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import pandas as pd
# 딥러닝용 모델링이므로 앞서 머신러닝시 수행단계 줄음
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df=pd.read_csv(data_url,sep="\s+",skiprows=22, header=None)
data=np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
# data=np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target=torch.FloatTensor(raw_df.values[1::2,2])
# target=raw_df.values[1::2,2]
x_data=torch.FloatTensor(data)
y_data=torch.FloatTensor(target)
y_data=y_data.view(506,1)
# torch.Size([506,13])
```

```python
class LinearRegressionModel(nn.Module):
  def __init__(self): #모델 정의 
    super().__init__()
    self.linear=nn.Linear(13,6)

    # 실행
  def forward(self,x):
    return self.linear(x)
```

```python
model=LinearRegressionModel()
optimizer=optim.SGD(model.parameters(),lr = 1e-6)
```

```python
for epoch in range(10000001):
  # 직선의 방정식에 해당하는 가설 
  # H(x) 계산 y=wx+b
  prediction=model(x_data) 
  # error의 제곱으로 손실
  cost=F.mse_loss(prediction,y_data)
  #gradient 0으로 초기화하기 안하면 미분 계속 누적됩니다 
  optimizer.zero_grad()
  # backward 적용연산 
  cost.backward()
  # w,b 업데이트
  optimizer.step()

  if epoch % 1000==0:
    print('Epoch=',epoch,'Cost=',cost.item())
    
  
```

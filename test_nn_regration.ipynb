```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
# 딥러닝용 모델링이므로 앞서 머신러닝시 수행단계 줄음
```

```python
torch.manual_seed(1)
# 변수 선언   
x_train=torch.FloatTensor([[73,80,75],[93,88,93],[89,91,90],[96,98,100,],[73,66,70]]).reshape(5,3)
y_train=torch.FloatTensor([[152],[185],[180],[196],[142]]).reshape(5,1)
```

```python
model=nn.Linear(3,1)
# input_dim, output_dim
print((model.parameters()))
```

```python
# SGD-확률적 경사하강법 lr 러닝데이터
optimizer=optim.SGD(model.parameters(),lr = 1e-5)
```

```python
for epoch in range(30001):
  # 직선의 방정식에 해당하는 가설 
  # H(x) 계산 y=wx+b
  predict=model(x_train) 
  # error의 제곱으로 손실
  cost=F.mse_loss(predict,y_train)
  #gradient 0으로 초기화하기 안하면 미분 계속 누적됩니다 
  optimizer.zero_grad()
  # backward 적용연산 
  cost.backward()
  # w,b 업데이트
  optimizer.step()

  if epoch % 100==0:
    print('Epoch=',epoch,'Cost=',cost.item())
  
```

```python
print(list(model.parameters()))
```

```python
class LinearRegressionModel(nn.Module):
  def __init__(self): #모델 정의 
    super().__init__()
    self.linear=nn.Linear(3,1)

    # 실행 
  def forward(self,x):
    return self.linear(x)
```

```python
model=LinearRegressionModel()
optim.SGD(model.parameters(), lr = 1e-5)
```

```python
for epoch in range(20001):
  predict=model(x_train) 
  cost=F.mse_loss(predict,y_train)

  optimizer.zero_grad()
  cost.backward()

  optimizer.step()

  if epoch % 400 ==0:
    print('Epoch',epoch,'cost',cost)
    
```
